```
(
  df
  .groupby('param_marka-pojazdu')['price_value'] # pogrupuj dane po marce pojazdu?
  .agg(np.mean) # zagreguj i wyciągnij średnią z price_value (['price_value'].agg(np.mean))
  .sort_values(ascending=False)  # posortuj malejąco
  .head(50) # ogranicz do najważniejszych 50
).plot(kind='bar', figsize=(20,5)) # wykres kind-rodzaj, figsize-rozmiar, za nawiasem ponieważ odnosi się to do całości
```

```
(
  df
  .groupby('param_marka-pojazdu')['price_value'] # pogrupuj dane po marce pojazdu?
  .agg(np.median) # zagreguj i wyciągnij medianę z price_value (['price_value'].agg(np.median))
  .sort_values(ascending=False)  # posortuj malejąco
  .head(50) # ogranicz do najważniejszych 50
).plot(kind='bar', figsize=(20,5)) # wykres kind-rodzaj, figsize-rozmiar, za nawiasem ponieważ odnosi się to do całości
```

```
(
  df
  .groupby('param_marka-pojazdu')['price_value'] # pogrupuj dane po marce pojazdu?
  .agg([np.mean, np.median, np.size]) # zagreguj i wyciągnij z price_value wszystkie 3 rzeczy
  .sort_values(by='size', ascending=False)  # posortuj malejąco, sortuj po średniej
  .head(50) # ogranicz do najważniejszych 50
).plot(kind='bar', figsize=(20,5), subplots=True) # wykres kind-rodzaj, figsize-rozmiar, za nawiasem ponieważ odnosi się to do całości
```

```
# Przepisanie na funkcję
def group_and_barplot(feat_groupby, feat_agg='price_value', agg_funcs=[np.mean, np.median, np.size], feat_sort='size', top=50, subplots=True):
  return(
    df
    .groupby(feat_groupby)[feat_agg]
    .agg(agg_funcs)
    .sort_values(by=feat_sort, ascending=False)
    .head(top)
  ).plot(kind='bar', figsize=(20,5), subplots=subplots)
```

```
group_and_barplot('param_kraj-pochodzenia');
```

```
group_and_barplot('param_kolor');
```
